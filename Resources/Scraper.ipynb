```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
URL='https://stats.espncricinfo.com/ci/engine/records/averages/batting.html?class=2;current=2;id=6;type=team'
source=requests.get(URL).text
```

```python
soup=BeautifulSoup(source,'lxml')
```

```python
head=soup.find('thead').text.split('\n')
head=[h for h in head if h is not '']
head
```

```python
def scrape(URL,filename):
    csv_file=open(filename,'w',newline='')
    csv_writer=csv.writer(csv_file)
    
    source=requests.get(URL).text
    soup=BeautifulSoup(source,'html5')
    
    head=soup.find('thead').text.split('\n')
    head=[h for h in head if h is not '' and h is not ' ']
    csv_writer.writerow(head)
    print(head)
    table=soup.find('tbody')
    for row in table.find_all('tr'):
        player=[]
        for t in row.find_all('td'):
            player.append(t.text)
            print(t.text, end=' ')
        csv_writer.writerow(player)
        print()
    csv_file.close()
```

```python
URL1='https://stats.espncricinfo.com/ci/engine/records/averages/batting.html?class=2;current=2;id=6;type=team'
filename1='batting_average.csv'
scrape(URL1,filename1)

URL2='https://stats.espncricinfo.com/ci/engine/records/averages/bowling.html?class=2;current=2;id=6;type=team'
filename2='bowling_average.csv'
scrape(URL2,filename2)
```
